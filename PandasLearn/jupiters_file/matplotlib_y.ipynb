```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Data for demonstration
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
data_scatter = pd.DataFrame({
    'x_val': np.random.rand(50) * 10,
    'y_val': np.random.rand(50) * 5 + 2 * np.random.rand(50),
    'category': np.random.choice(['A', 'B', 'C'], 50)
})
print(data_scatter)
```

```python
# --- Example 1: Basic Line Plot ---
plt.figure(figsize=(10, 6)) # Create a new figure and set its size
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-.')
plt.plot(x, y2, label='cos(x)', color='red', linestyle=':')
# Adding labels and title
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Sine and Cosine Waves")
plt.legend() # Display the legend for labels
plt.grid(True) # Add a grid for better readability
plt.show() # Display the plot
```

```python
# --- Example 2: Basic Scatter Plot ---
plt.figure(figsize=(8, 6))
plt.scatter(data_scatter['x_val'], data_scatter['y_val'],
            alpha=0.7, # Transparency
            c=data_scatter['y_val'], # Color points based on y_val
            cmap='Set1', # Colormap
            s=data_scatter['x_val'] * 10 # Size points based on x_val
           )
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Random Scatter Plot with Color and Size Variation")
plt.colorbar(label="Y Value") # Add a color bar
plt.show()
```

```python
# --- Example 3: Bar Chart (using a Pandas Series) ---
fruit_counts = pd.Series({'Apples': 25, 'Oranges': 15, 'Bananas': 30, 'Grapes': 10})
plt.figure(figsize=(7, 5))
fruit_counts.plot(kind='bar', color='skyblue')
plt.title("Fruit Counts")
plt.xlabel("Fruit Type")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels
plt.tight_layout() # Adjust plot to prevent labels from overlapping
plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
np.random.seed(42)
data_dist = pd.Series(np.random.normal(loc=10, scale=2, size=1000))
```

```python
# --- Example 1: Distribution Plot (Histogram + KDE) ---
plt.figure(figsize=(9, 5))
sns.histplot(data_dist, kde=True, bins=30, color='purple')
plt.title("Distribution of Sample Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# --- Example 2: Scatter Plot with a Third Variable (Hue) ---
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="time", size="size", data=tips, palette="viridis", s=100)
sns.regplot(x="total_bill", y="tip", data=tips, logx=True)
plt.axvline(x=tips['total_bill'].median(), color='red', linestyle='--')
plt.axvline(x=tips['total_bill'].mode().tolist(), color='orange', linestyle='--')
plt.axvline(x=tips['total_bill'].mean(), color='violet', linestyle='--')
plt.axvline(x=tips['total_bill'].std(), color='b', linestyle='--')
plt.title("Total Bill vs. Tip by Time of Day and Party Size")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
# --- Example 3: Bar Plot (Categorical vs. Numerical) ---
plt.figure(figsize=(8, 5))
sns.barplot(x="day", y="total_bill", hue="day", data=tips, palette="pastel", legend=False)
plt.title("Average Total Bill by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
# --- Example 4: Box Plot (Comparing Distributions) ---
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="coolwarm")
plt.title("Total Bill Distribution by Day and Smoker Status")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.legend(title="Smoker")
plt.show()
```

```python
# --- Example 5: Pair Plot (Visualizing relationships across the entire DataFrame) ---
# Creates a grid of scatter plots for each pair of numerical variables
# and histograms/KDEs for each single numerical variable.
sns.pairplot(tips, hue="smoker", palette="rocket")
plt.suptitle("Pair Plot of Tips Dataset by Smoker Status", y=1.02) # Adjust suptitle position
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Create a Seaborn plot
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips, palette="dark:salmon_r")

# Now, use Matplotlib functions to enhance it
plt.title("Enhanced Scatter Plot with Matplotlib Customizations")
plt.xlabel("Total Bill Amount ($)", fontsize=12, color='darkgreen')
plt.ylabel("Tip Amount ($)", fontsize=12, color='darkgreen')
plt.xticks(rotation=45)
plt.axvline(x=25, color='gray', linestyle='--', label='Threshold') # Add a vertical line
plt.text(26, 8, 'High Bills', color='gray') # Add some text
plt.legend(loc='upper left')
plt.grid(axis='y', linestyle=':', alpha=0.7) # Add a horizontal grid
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x**2)
y4 = np.exp(-x/2) * np.sin(x)

# Create a figure and a 2x2 grid of subplots

# Subplot 1 (top-left)
plt.subplot(2, 2, 1) # 2 rows, 2 columns, 1st subplot
plt.plot(x, y1, 'r')
plt.title('Sine Wave')

# Subplot 2 (top-right)
plt.subplot(2, 2, 2) # 2 rows, 2 columns, 2nd subplot
plt.plot(x, y2, 'g--')
plt.title('Cosine Wave')

# Subplot 3 (bottom-left)
plt.subplot(2, 2, 3) # 2 rows, 2 columns, 3rd subplot
plt.plot(x, y3, 'b-.')
plt.title('Sine of x^2')

# Subplot 4 (bottom-right)
plt.subplot(4, 4, 15) # 2 rows, 2 columns, 4th subplot
plt.plot(x, y4, 'k:')
plt.title('Damped Sine Wave')

# Adjust layout to prevent titles/labels from overlapping
plt.tight_layout()

# Display the figure with all subplots
plt.show()
```
